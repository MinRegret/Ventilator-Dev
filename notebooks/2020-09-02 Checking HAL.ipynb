```python
import time
import datetime
import tqdm
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = []
ts = []

hal.setpoint_in = 10

for i in tqdm.tqdm(range(200)):
    if i % 40 == 0:
        hal.setpoint_in += 10
    xs.append(hal.pressure)
    ts.append(datetime.datetime.now().timestamp())
    time.sleep(0.1)

hal.setpoint_in = 0
ts, xs = np.array(ts), np.array(xs)
```
